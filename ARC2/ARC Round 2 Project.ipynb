```python
import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
 #Might take time to initialize. Also ignore 'unsaved custom gradient warning'
hrnet_model_name = 'camvid-hrnetv2-w48/1'   
tfhub_model_name = 'https://tfhub.dev/google/HRNet/' + hrnet_model_name
hrnet_model = hub.load(tfhub_model_name)
```

```python
#Path to image
img=np.array(Image.open(r"C:\Users\Asus\Desktop\ARC2\Sample_1.png"))/255.0
```

```python
#To check if image has been properly loaded
predictions = hrnet_model.predict([img])
plt.imshow(img)
plt.title("Original")
plt.show()
        
```

```python
#Plain white image means that the particular class is not present in the picture
#Road
    
#Create a mask containing required region by using the parts focussed by the final layer output
mask = predictions[0, :, :, 18] > 0.5      #4th dimension is the class number

# Initialize a white image
result_img = np.ones_like(img) * 255 

# Copy the regions where mask is True from the original image
result_img[mask] = img[mask]
plt.title("Road")
plt.imshow(result_img)
```

```python
#Human
mask = predictions[0, :, :, 17] > 0.5     
result_img = np.ones_like(img) * 255 
result_img[mask] = img[mask]
plt.title("Human")
plt.imshow(result_img)
```

```python
#Motor Vehicles

#Original model has multiple classes for different types of motor vehicles
mask1 = predictions[0, :, :, 6] > 0.5    
mask2 = predictions[0, :, :, 23] > 0.5
mask3 = predictions[0, :, :, 28] > 0.5
mask4=mask1|mask2|mask3
result_img = np.ones_like(img) * 255 
result_img[mask4] = img[mask4]
plt.title("Motor Vehicles")
plt.imshow(result_img)
```
